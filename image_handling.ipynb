```python
#STEP 6 (a)
#Converting all images to grey data set
grayImg=[]
for i in images:
  newImg = i.convert('L')
  #grayImg = cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)
  grayImg.append(newImg)
grayImg[2]
```

```python
#STEP 6 (b)
#converting all images to binary
BinaryImg=[]
for i in images:
  BImg = i.convert('1')
  BinaryImg.append(BImg)
BinaryImg[1]
```

```python
#STEP 6 (c)
#flattening the image matrix
FlattenImg=[]
for i in images:
  arr1=np.array(i)
  arr2=arr1.flatten()
  FlattenImg.append(arr2)
FlattenImg[1]  
```
